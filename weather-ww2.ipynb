```python
import pandas as pd
import os

WEATHER_PATH = "datasets/weather"

def load_weather_data(weather_path=WEATHER_PATH):
    csv_path = os.path.join(weather_path, "Summary_of_Weather.csv")
    return pd.read_csv(csv_path)

weather = load_weather_data()
weather.head()
```
